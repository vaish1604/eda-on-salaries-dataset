Through Feature engineering we can say that salary_in_usd is more uniform basis of comparison. Hence, we will drop the columns salary and salary_currency

```python
data=data.drop(['salary','salary_currency'],axis=1)
data.head()
```

We will now look at a few statistics for the data

```python
data.describe().T
```

We have noticed that some of columns include non-numeric values (experience_level,employment_type,employee_residence,company_location,company_size)

We will convert these non numeric values to numeric so that we can include them in our analysis

We call this process encoding, there are multiple ways of encoding, for this project, we will be using Label Encoding

```python
from sklearn import preprocessing
data1=data
label_encoder = preprocessing.LabelEncoder()
data1['experience_level'] = label_encoder.fit_transform(data['experience_level'])
data1['employment_type'] = label_encoder.fit_transform(data['employment_type'])
data1['employee_residence'] = label_encoder.fit_transform(data['employee_residence'])
data1['company_location'] = label_encoder.fit_transform(data['company_location'])
data1['company_size'] = label_encoder.fit_transform(data['company_size'])
data1.head()
```

If we look at the data.describe() function again it will include all the columns now

```python
data1.describe().T
```

Going column by column to check its dependency on the data is a long and tedious task.
In our case we have only 10 columns, but a lot of datasets have multiple features (columns) which make it very difficult to draw conclusions.

To counter this, we have a lot of methods to check the impact of columns on each other, I have used a correlation matrix.

```python
print(data1.corr())
```

```python
plot1=sns.heatmap(data1.corr(),cmap="YlGnBu",annot=True) #using heatmap to plot the correlation ,matrix
plt.show()
```

From the above correlation matrix, if we look at only the salary_in_usd row or column, employee_residence and company_location have maximum impact, followed by experience_level and work_year

On the other hand, company_size has minimum correlation to salary_in_usd

Let us now plot line graphs for each significant feature against the salary

```python
fig, axs = plt.subplots(2,2)
axs[0,0].bar(data['employee_residence'],data['salary_in_usd'])
axs[0,0].set_title("employee_residence")
axs[0,1].bar(data['company_location'],data['salary_in_usd'])
axs[0,1].set_title("company_location")
axs[1,0].bar(data['experience_level'],data['salary_in_usd'])
axs[1,0].set_title("experience_level")
axs[1,1].bar(data['work_year'],data['salary_in_usd'])
axs[1,1].set_title("work_year")
```
